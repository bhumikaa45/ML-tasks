```python
Q5. Explain the difference between a list, tuple, set, and dictionary in Python.

Answer- The main difference among them lies in mutability (whether you can change them) and order. 
Lists and Dictionaries are changeable, while Tuples are permanent once created, and Sets are unique collections that don't allow duplicates.

Data Structure Ordered  Mutable  Duplicates Allowed 

 List           Yes      Yes      Yes                
 Tuple          Yes      No       Yes                
 Set            No       Yes      No                 
 Dictionary     Yes      Yes      Keys unique        

1. List []
    The "all-purpose" container. Use it when you have a collection of items where the order matters and you need to add or remove things later.
    Example: shopping_list = ["apples", "bread", "milk"]

2. Tuple ()
    A "locked" list. Use it for data that should not change (like coordinates or settings). They are faster and safer than lists because they are immutable.
    Example: point = (10, 20, 30)

3. Set {}
    An unordered collection of unique items. Use it to remove duplicates from a list or to perform math operations like unions and intersections.
    Example: unique_ids = {101, 102, 103}

4. Dictionary {"key": "value"}
    A map of key-value pairs. Use it when you want to look up a value using a specific label instead of an index number.
    Example: user_info = {"name": "Alice", "age": 25}
```

```python
#Q6. Write a Python function that takes a list of numbers and returns only the even numbers.

#Answer- 
def get_even_numbers(numbers):
    return [num for num in numbers if num % 2 == 0]

numbers = [1,2,3,4,5,6,7,8]
print(get_even_numbers(numbers))
```

```python
'''Q7. What is the difference between == and is in Python?

Answer- The fundamental difference is that == checks for equality (value), while is checks for identity (memory location).
        ~ == (Equality): Evaluates to True if the values of the two objects are the same.
        ~ is (Identity): Evaluates to True only if both variables point to the exact same object in memory.'''

a = [1,2]
b = [1,2]

print(a == b)
print(a is b)
```

```python
'''Q8. What is the use of conditional statements in Python?

Answer- With conditional statements, the code can choose to do something depending on the conditions that are either true or false. 
Without conditional statements, programs will just blindly follow all commands; with conditional statements, the code can actually have options based on certain data encountered.

They rely on the if, elif (else if), and else keywords:

if: The starting point. If this condition is true, run the code inside.
elif: Checked only if the previous conditions were false. You can have many of these.
else: The "catch-all." If nothing above was true, run this.'''

age = 20

if age >= 18:
    print("Adult")
elif age > 12:
    print("Teenager")
else:
    print("Child")
```

```python
Q33. Explain the difference between independent and dependent events with one example each.

Answer- Independent:-
These are events where the outcome of the first has zero effect on the second. They are "memoryless."
The Logic: P(A and B) = P(A)*P(B)
Example: Tossing a coin. If you flip "Heads" on the first try, the chance of getting "Heads" on the second try is still exactly 50%. The coin doesn't "remember" the first flip.

Dependent Events:-
These are events where the outcome of the first changes the probability of what happens next.
The Logic: The "pool" of possible outcomes shrinks or shifts after the first event.
Example:

Drawing two cards without replacement.
The first card drawn affects the probability of the second card.

 Independent Events                    Dependent Events            

 One event does not affect the other   One event affects the other 
 Probabilities remain unchanged        Probabilities change        
```

```python
#Q34. A coin is tossed twice. List all possible outcomes and calculate the probability of getting exactly one head.

#Answer-

from itertools import product

outcomes = list(product(['H', 'T'], repeat=2))
print("All Outcomes:")
print(outcomes)

# Exactly one head
one_head = [outcome for outcome in outcomes if outcome.count('H') == 1]

print("\nOutcomes with exactly one head:")
print(one_head)
probability = len(one_head) / len(outcomes)

print("\nProbability =", probability)
```

```python
Q35. What is conditional probability? Explain with a simple example.

Answer- Conditional probability is the probability of an event occurring given that another event has already occurred.

Formula:
        P(A\B) = P(A (union) B)/ P(B)

Example:-

A class has:

20 boys
30 girls

If a selected student is known to be a girl, the probability of choosing a girl is:

P(Girl∣Student is Girl)=1

Conditional probability updates probability using additional information.
```
