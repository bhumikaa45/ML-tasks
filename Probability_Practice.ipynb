```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from scipy.stats import norm, binom
```

```python
Coin toss Simulation
```

```python
outcomes= ['Heads','Tails']
result= [random.choice(outcomes) for _ in range(1000)]
HP= result.count('Heads')/len(result)
TP= result.count('Tails')/len(result)
print("P(Heads):",HP)
print("P(Tails):",TP)
```

```python
Dice Roll Distribution
```

```python
rolls= np.random.randint(1,7,1000)
plt.hist(rolls, bins=6,edgecolor='pink',color='brown')
plt.title("Dice Roll")
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.show()
```

```python
Condtitional Probability 
```

```python
data={'Customer':['A','B','C','D','E'],
        'Churn':['Yes','No','Yes','No','Yes'],
        'Plan':['Premium','Basic','Premium','Basic','Premium']}
df=pd.DataFrame(data)
premium_customers=df[df['Plan']=='Premium']
prob_churn_given_premium=(premium_customers['Churn']=='Yes').mean()
print("P(Churn|Premium):",prob_churn_given_premium)
```

```python
Binomial Distribution
```

```python
n,p= 10,0.5
x=np.arange(0,n+1)
pmf=binom.pmf(x,n,p)
plt.bar(x,pmf,color='cyan')
plt.title("Binomial Distribution(10 tosses p=0.5)")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.show()
```

```python
Normal Distribution
```

```python
x=np.linspace(-3,3,100)
plt.plot(x,norm.pdf(x,0,1),color='magenta')
plt.title("Normal Distribution(mean=0,std=1)")
plt.xlabel("x")
plt.ylabel("Probability density")
plt.show()
```
