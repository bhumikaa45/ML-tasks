```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
GA = [23,21,19,24,30,22,25]
GB = [28,27,29,31,26,30,32]
print("Group A Mean:",np.mean(GA))
print("Group B Mean:",np.mean(GB))
```

```python
Perform t-test
```

```python
t_stat,p_value= stats.ttest_ind(GA,GB)
print("t-statistic", t_stat)
print("p-value", p_value)
```

```python
Confidence Interval
```

```python
mean_diff=np.mean(GA)-np.mean(GB)
print("Mean Difference:", mean_diff)
```

```python
plt.boxplot([GA,GB], labels=['Group A','Group B'])
plt.title("Group Comparison")
plt.show()
```
