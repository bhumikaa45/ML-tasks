```python
Find the second largest element in a list without using sorted()
```

```python
nums= [10, 20, 4, 45, 99]
first= second= float('-inf')
for n in nums:
    if n>first:
        second = first
        first = n
    elif n>second and n!= first:
        second= n
print(second)
```

```python
Find the common element between the two lists
```

```python
l= [1,2,3,4,5]
m=[6,7,5,8,1]
print([x for x in l if x in m])
```

```python
Flatten a nested list into a single list.
```

```python
nested= [[1,2],[3,4],[5]]
flat= []
for sub in nested:
    for x in sub:
        flat.append(x)
print(flat)
```

```python
Merge two dictonaries and sum values of the common keys
```

```python
d1= {"a":2,"b":3}
d2= {"b":4,"c":5}
merged= d1.copy()
for k,v in d2.items():
    merged[k]= merged.get(k,0)+v
print(merged)
```

```python
Sort dictionary by values in descending order.
```

```python
scores= {"A":50,"B":80,"C":40}
sorted_scores= {k:v for k,v in sorted(scores.items(),key=lambda x:x[1],reverse=True)}
print(sorted_scores)
```

```python
Factorial (Recursive)
```

```python
def factorial(n):
    return 1 if n==0 else n*factorial(n-1)
print(factorial(5))
```

```python
Prime Check Function
```

```python
def is_prime(n):
    if n<2: return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0: return False
    return True
print(is_prime(17))
```

```python
Fibonacci Function
```

```python
def fibonacci(n):
    a,b=0,1
    seq=[]
    for _ in range(n):
        seq.append(a)
        a,b=b,a+b
    return seq
print(fibonacci(10))  
```

```python
Palindrome Function
```

```python
def is_palindrome(s):
    return s==s[::-1]
print(is_palindrome("Elephant"))
```

```python
 Armstrong Numbers (1–500)
```

```python
for num in range(1,501):
    digits=str(num)
    total=sum(int(d)**len(digits) for d in digits)
    if total==num:
        print(num,end=" ")
```
