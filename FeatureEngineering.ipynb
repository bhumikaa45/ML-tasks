```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("cleaned_model_ready.csv")
print(df.head())
```

```python
df["sales_to_demand_ratio"]=df["sales_qty"]/df["target_demand_qty"]
```

```python
df['sales_lag_1m']=df['sales_qty'].shift(1)
df['sales_lag_2m']=df['sales_qty'].shift(2)
```

```python
df['rolling_3m_sales']=df['sales_qty'].rolling(window=3).mean()
df['rolling_3m_demand']=df['target_demand_qty'].rolling(window=3).mean()
```

```python
df['supplier_activity'] = df['n_suppliers'] * df['service_qty']
```

```python
df.to_csv("feature_engineered_dataset.csv", index=False)
```
