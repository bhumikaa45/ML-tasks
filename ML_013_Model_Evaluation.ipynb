```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")

y_train = pd.read_csv("y_train.csv")
y_test = pd.read_csv("y_test.csv")
```

```python
y_train = y_train.squeeze()
y_test = y_test.squeeze()

print(type(y_train))
print(type(y_test))
```

```python
y_train = y_train.squeeze()
y_test = y_test.squeeze()

print(type(y_train))
print(type(y_test))
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
print("Model trained successfully")
```

```python
y_pred = model.predict(X_test)
print("First 5 Predictions:")
print(y_pred[:5])
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)
print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
# Model Evaluation Metrics

MAE (Mean Absolute Error):
Measures the average prediction error.

MSE (Mean Squared Error):
Measures the average squared error.

RMSE (Root Mean Squared Error):
Represents prediction error in the same unit as the target variable.

R² Score:
Indicates how much variance in the target variable is explained by the model.
A value closer to 1 indicates better performance.
```
