```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
A dataset
```

```python
data={
    'Age': [25, 30, 22, 28, 35],
    'Income': [4000, 5000, 3500, 4500, 6000],
    'Spending': [2000, 2500, 1500, 2200, 3000]
}
df=pd.DataFrame(data)
print(df)
```

```python
Covariance
```

```python
print("Covariance Matrix:\n",df.cov())
```

```python
Correlation
```

```python
print("Correlation Matrix:\n", df.corr())
```

```python
Correlation Heatmap
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
Scatter Plot
```

```python
plt.scatter(df['Income'],df['Spending'],color='seagreen')
plt.title("Income vs Spending")
plt.xlabel("Income")
plt.ylabel("Spending")
plt.show()
```
