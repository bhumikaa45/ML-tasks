```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
print(df.shape)
df.head()
```

```python
df['TotalCharges'] = pd.to_numeric(
    df['TotalCharges'],
    errors='coerce'
)

df['TotalCharges'] = df['TotalCharges'].fillna(
    df['TotalCharges'].median()
)

df.drop('customerID', axis=1, inplace=True)

df['Churn'] = df['Churn'].map({
    'No': 0,
    'Yes': 1
})

df = pd.get_dummies(
    df,
    drop_first=True
)
print(df.shape)
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
print(X_train.shape)
print(X_test.shape)
```

```python
model = DecisionTreeClassifier(
    random_state=42
)
model.fit(X_train, y_train)
print("Decision Tree trained successfully")
```

```python
y_pred = model.predict(X_test)
print(y_pred[:10])
```

```python
accuracy = accuracy_score(
    y_test,
    y_pred
)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(
    y_test,
    y_pred
)
print(cm)
```

```python
print(
    classification_report(
        y_test,
        y_pred
    )
)
```

```python
importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': model.feature_importances_
})
importance = importance.sort_values(
    by='Importance',
    ascending=False
)
print(importance.head(10))
```

# Conclusion

A Decision Tree Classifier was built to predict customer churn.

The dataset was preprocessed by handling missing values, encoding categorical variables, and splitting the data into training and testing datasets.

The Decision Tree model was trained and evaluated using Accuracy Score, Confusion Matrix, and Classification Report.

Feature importance analysis was also performed to identify the most influential factors affecting customer churn.
