```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
Line chart
```

```python
data = {'Month':['Jan','Feb','Mar','Apr','May'],
        'Sales':[200, 250, 300, 280, 350]}
df = pd.DataFrame(data)
plt.plot(df['Month'],df['Sales'],marker='o',color='purple')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
Bar chart
```

```python
plt.bar(df['Month'],df['Sales'],color='pink')
plt.title('Monthly Sales Bar Chart')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
Scatter Plot
```

```python
plt.scatter(df['Month'],df['Sales'],color='brown')
plt.title('Monthly Sales Scatter Plot')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
Histogram
```

```python
plt.hist(df['Sales'],bins=5,color='blue',edgecolor='black')
plt.title('Sales Distribution Histogram')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
Pie Chart
```

```python
plt.pie(df['Sales'],labels=df['Month'],autopct='%1.1f%%')
plt.title('Sales Share by Month')
plt.show()
```
