```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Create Dataset
```

```python
data= {'Customer': ['A','B','C','D','E'],
        'Age': [25, 30, 22, 28, 35],
        'Purchase amount': [200, 250, 300, 280, 350]}
df= pd.DataFrame(data)
print(df)
```

```python
Statistics calculations
```

```python
print("Mean age:",df['Age'].mean())
print("Median age:",df['Age'].median())
print("Mode age:",df['Age'].mode()[0])
print("Variance age:",df['Age'].var())
print("Standard Deviation age:",df['Age'].std())
```

```python
Descirbe entire dataset
```

```python
print(df.describe())
```

```python
Visual Distribution
```

```python
plt.hist(df['Purchase amount'], bins=5, color='pink', edgecolor='maroon')
plt.title('Purchase amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```
