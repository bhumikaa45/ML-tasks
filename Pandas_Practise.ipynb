```python
Read CSV file
```

```python
import pandas as pd
#Example 
data= {
    'Name': ['A', 'B', 'C', 'D'],
    'Age': [25, 30, 22, 28],
    'Department': ['HR', 'IT', 'Finance', 'Marketing'],
    'Score': [80, 90, 70, 85]
}
#Convert to Dataframe
df= pd.DataFrame(data)
# Save as CSV file
df.to_csv("example.csv", index=False)
print("example.csv file created successfully!")
```

```python
df= pd.read_csv("example.csv")
print(df.head())
```

```python
Filtering, Slicing
```

```python
print(df[df['Age'] > 25])
print(df[['Name','Score']])
```

```python
GroupBy, Aggregation
```

```python
print(df.groupby('Department')['Score'].mean())
```

```python
Merge
```

```python
df1= pd.DataFrame({'ID':[1,2],'Name':['A','B']})
df2= pd.DataFrame({'ID':[1,2],'Salary':[5000,6000]})
mer= pd.merge(df1,df2,on='ID')
print(mer)
```
